## Question 3

$$
elu(x) = 
\left\{
\begin{array}{r l}
x        & \text{if } x \geq 0 \\
\alpha(e^x-1) & \text{else}
\end{array}
\right.
$$

```python
def elu(z,alpha=1):
    new_z = z.copy()
    new_z[z<0] = alpha*(np.exp(z[z<0])-1)
    return new_z
```

```python
def relu(z):
    new_z = z.copy()
    new_z[z<0] = 0
    return new_z
```

```python
plt.figure(figsize=(13,3))
plt.subplot(121)
x = np.linspace(-3,3,100)
y = elu(x)
plt.plot(x,y)
plt.title("elu")


plt.subplot(122)
x = np.linspace(-3,3,100)
y = relu(x)
plt.plot(x,y)
plt.title("relu")
```

Like rectified linear units (ReLUs), leaky ReLUs (LReLUs) and parametrized ReLUs (PReLUs), ELUs alleviate the vanishing gradient problem via the identity for positive values. However, ELUs have improved learning characteristics compared to the units with other activation functions. In contrast to ReLUs, ELUs have negative values which allows them to push mean unit activations closer to zero like batch normalization but with lower computational complexity. (cf paper from Djork-Arné Clevert, Thomas Unterthiner, Sepp Hochreiter)
